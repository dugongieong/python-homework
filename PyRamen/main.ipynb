````python
# Importing cvs's for PyRamen.

# Import Libraries
import csv
from pathlib import Path

# Set path for sales_data.csv & menu_data.csv to collect data
menu_csvpath = Path('Resources', 'menu_data.csv') 
sales_csvpath = Path('Resources', 'sales_data.csv')

# Initilize an empty dict or lists objects to hold content of csv files.
menu = []
sales = []
report = {}
row_count = 0

# Open/Read Budget_Data.csv as an object menu csv
with open(menu_csvpath, 'r') as csvfile:
    
    # Skip the header (first row of the CSV)
    csv_reader = csv.reader(csvfile, delimiter = ',')
    header = next(csv_reader)
        
    # Loop over the rest of the rows and append every row to the `menu` list object
    for row in csv_reader:
        menu.append(row)

# Open/Read Budget_Data.csv as an object sales csv
with open(sales_csvpath, 'r') as csvfile:
        
    # Skip the header (first row of the CSV)
    csv_reader = csv.reader(csvfile, delimiter = ',')
    header = next(csv_reader)
    
    for row in csv_reader:
        sales.append(row)

# append every row of the sales data to a new `sales` list object
for sales_row in sales:
    sales_item = sales_row[4] 
    quantity = sales_row[3]

    #set the `sales_item` as a new key to the `report` dictionary and the values as a nested dictionary containing the following:
    if sales_item not in report:
        report[sales_item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0,}
            
        
    else:
        continue

# Setting values for sales to report
for sales_row in sales:
    
    sales_item = sales_row[4]
    quantity = sales_row[3]
    
    
    # Setting values for menu to report
    for menu_row in menu:
        menu_item = menu_row[0]
        menu_price = menu_row[3]
        menu_cost = menu_row[4]
        
        
#sales_item` in sales is equal to the `item` in `menu`, capture the `quantity`        
#```python
#report[sales_item]["01-count"] += quantity
#report[sales_item]["02-revenue"] += price * quantity
#report[sales_item]["03-cogs"] += cost * quantity
#report[sales_item]["04-profit"] += profit * quantity        
# Else print the message "{sales_item} does not equal {item}! NO MATCH!".        
        if sales_item == menu_item:
            report[sales_item]["01-count"] += (int(quantity))
            report[sales_item]["02-revenue"] += (int(menu_price) * int(quantity))
            if menu_item == sales_item:
                report[menu_item]["03-cogs"] += (int(menu_cost) * int(sales_row[3]))
                report[menu_item]["04-profit"] += (int(menu_price) * int(quantity)) - (int(menu_cost) * int(sales_row[3]))

                
# output file with metrics
output_path = Path("PyRamen.txt")

with open(output_path, 'w') as file:
    file.write("This is the financial report for PyRamen.\n")
    for data in report:
        file.write(f"{data} {report[data]} \n")
````
